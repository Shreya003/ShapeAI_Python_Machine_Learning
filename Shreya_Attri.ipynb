```python
import numpy as np
import pandas as pd
import sklearn 
from sklearn.datasets import load_boston
data1 = load_boston()
data1.keys()
print(data1.DESCR)
print(data1.data)
boston_tabular = pd.DataFrame(data1.data, columns=data1.feature_names)
boston_tabular.head(6)
boston_tabular['Output'] = data1.target
boston_tabular.head()
boston_tabular.isnull()
boston_tabular.isnull().sum()

from sklearn.model_selection import train_test_split

x = boston_tabular.drop('Output', axis=1)
y = boston_tabular['Output']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state=4)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lin_model = LinearRegression()
lin_model.fit(x_train, y_train)

y_train_predict = lin_model.predict(x_train)
rmse = (np.sqrt(mean_squared_error(y_train, y_train_predict)))

print("The performance of the training set")
print('RMSE is {}'.format(rmse))
print("\n")

y_test_predict = lin_model.predict(x_test)
rmse = (np.sqrt(mean_squared_error(y_test, y_test_predict)))

print("The performance for testing set")
print('RMSE is {}'.format(rmse))

                




```
